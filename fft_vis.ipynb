```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import distinctipy
```

```python
N = 8

X = []
print(X)

for k in range(N):
	X_k = []
	for n in range(N):
		# X_k.append(f"x{n}.e^(-2.pi.i.{k}.{n}/{N})")
		val = f"np.e^(-2*np.pi*1j*{k}*{n}/{N})"
		val = np.power(np.e,-2*np.pi*1j*k*n/N)
		# print(np.power(np.e,-2*np.pi*1j*k*n/N))
		X_k.append(val)
		
		# if val not in colors.keys():
		# 	color_count += 1
		# 	colors[val] = 

	X.append(X_k)

df = pd.DataFrame(X, columns=[f"x{i}" for i in range(N)], index=[f"X{i}" for i in range(N)])
print(df)
```

```python
color_map = dict(zip(pd.unique(df.to_numpy().flatten()), 
				[distinctipy.get_hex(i) for i in distinctipy.get_colors(pd.unique(df.to_numpy().flatten()).size)]
			))

cell_color = []
color_num = []
for i in range(df.to_numpy().shape[0]):
	cell_color_row = []
	color_num_row = []
	for j in range(df.to_numpy().shape[1]):
		cell_color_row.append(color_map[df.to_numpy()[i][j]])
		color_num_row.append(list(color_map.keys()).index(df.to_numpy()[i][j]))
	cell_color.append(cell_color_row)
	color_num.append(color_num_row)

fig, ax = plt.subplots()
ax.axis('tight')
ax.axis('off')
the_table = ax.table(cellText=color_num,cellColours=cell_color,loc='center')
```
